**Pandas Series** pandas series is a 1-d laeled array that can holds any type of data (Hetrogeneous Data)

```python
import pandas as pd
# s = pd.Series(data,index = index) here data can be a list dict tuple numpy array
s = pd.Series([14,24,34,54,44])
print(s)

names = pd.Series([20,30,40,50], index = ["mukul","neelam", "rajveer", "neeraj"])
print(names)

data = {"Amit": 85, "Priya": 90, "Rahul": 78, "Sneha": 92}
marks = pd.Series(data)
print(marks)
```

```python
print(names['mukul'])
print(names[["mukul","neelam"]])
```

**pandas Vectorized operations**  -- no need to iterate over a loop pandas apply it to every element autometically

```python
print(s.mean())
print(s.min())
print(s.max())
print(s.sum())
print(s>30)
print(s.diff())
```

**Series "name" attribute** -- it is useful when we combine the multiple series into a dataframe, each series become column and name become the column header

```python
marks = pd.Series([10,20,30,40], name ="Marks")
names = pd.Series(["mukul","neelam","rajveer","neeraj"], name="Students Name")
df1 = pd.concat([marks,names], axis=1)
print(df1)
```

**Accessing in pandas dataframes**
1. df["col"] -- it returns a column
2. df.loc[] --- labeled based (names of row/columns)
3. df.iloc[] --- position based (position of row/column)

```python
data = {
    "name": ['a','b','c','d','e'],
    "marks" : [10,20,30,40,50],
   "grade" : [1,2,3,4,5],
    "result" : ['p','f','f','p','p']

}
df =pd.DataFrame(data)
print(df)
```

```python
df.shape
```

```python
print(df["marks"])
print(df[["marks","grade"]])
```

```python
print(df.loc[0],end='\n \n')
print(df.loc[0:2],end='\n\n')
print(df.loc[0,"name"],end='\n\n')
print(df.loc[0:3, "name": "grade"])
```

```python
print(df.iloc[0])
print(df.iloc[0:2])
print(df.iloc[0,2])
print(df.iloc[0:2,1:4])
```

the main difference loc and iloc is that loc is labeled based and end inclusive and ilocis no based and end exclusive

**Filtering in pandas**

```python
# find studens whose score >85
print(df["Score"]>85) # returns the series of true and false (Boolean Series)
df1=pd.DataFrame(df["Score"]>85) # return the dataframe of boolean series

print(df1)
df2=df[df["Score"]>85] # uses boolean series as filter and return where condition is true
df2
```

```python
#multiple conditions
df3 = df[(df['Score'] > 80) & (df['Age'] < 25)]
print(df3)

df3 = df[(df['Score'] > 80) | (df['Age'] < 25)]
print(df3)

df[(df['City'] == 'Delhi')]
```

```python
print(df[df['City'].isin(['Delhi', 'Mumbai'])])
print(df[df['Score'].between(80, 90)])
```

**Vectorization** --- Performing an operation on an entire Series/DataFrame at once instead of processing one row at a time.

```python
df = pd.DataFrame({
    "Name":["A","B","C","D"],
    "Salary":[50000,60000,40000,55000]
})
print(df)
```

```python
new_salary = [] #non-vectorized approach
for salary in df["Salary"]:
    new_salary.append(salary + 5000)
df["New_Salary"] = new_salary
df
```

```python
df["New_Salary"] = df["Salary"] + 5000 # vectorized approach
df
```

**groupby()** splits data into groups based on a column and then performs calculations on each group.

```python
df = pd.DataFrame({
    "Department":["IT","IT","HR","HR","Sales"],
    "Employee":["A","B","C","D","E"],
    "Salary":[50000,70000,40000,60000,55000]
})

df
```

```python
df.groupby("Department")["Salary"].mean()
```

at first it group the data by department and them apply mean on it

```python
df.groupby("Department")["Salary"].sum()
```

```python
df.groupby("Department")["Salary"].agg(["mean","sum","max","min","count"])
```

```python
df = pd.DataFrame({
    "Department":["IT","IT","HR","HR"],
    "Gender":["M","F","M","F"],
    "Salary":[50000,70000,40000,60000]
})
```

```python
pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="mean"
)
```

A Pivot Table is a tool used to summarize and aggregate data by grouping rows and columns and applying calculations such as:,Sum,Mean,Count,Min,Max
It helps convert detailed data into a summary report.

```python
pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns="Gender",
    aggfunc="mean"
)
```
